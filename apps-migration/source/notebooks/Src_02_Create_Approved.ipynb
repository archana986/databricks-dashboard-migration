# Apps Migration: Create Approved Inventory

This notebook copies the inventory to the approved location for demo purposes.

## Configuration

```python
try:
    catalog = dbutils.widgets.get("catalog")
except:
    catalog = "YOUR_SOURCE_CATALOG"

try:
    schema = dbutils.widgets.get("schema")
except:
    schema = "apps_demo"

try:
    volume = dbutils.widgets.get("volume")
except:
    volume = "apps_migration"

volume_path = f"/Volumes/{catalog}/{schema}/{volume}"
print(f"Volume path: {volume_path}")
```

## Copy inventory to approved

```python
from databricks.sdk import WorkspaceClient
from io import BytesIO

w = WorkspaceClient()

# Read inventory
inventory_path = f"{volume_path}/apps_inventory/apps_inventory.csv"
response = w.files.download(inventory_path)
content = response.contents.read()

print(f"Read inventory: {len(content)} bytes")

# Write to approved
approved_path = f"{volume_path}/apps_inventory_approved/inventory_approved.csv"
w.files.upload(approved_path, BytesIO(content), overwrite=True)

print(f"Approved inventory created: {approved_path}")
```
